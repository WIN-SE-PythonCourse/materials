## Reading Corner - Part 1: if-else Branches

This reading corner again uses the simple example of calculating sales tax and gross amount to show how branches are used in Python. First, read the algorithm in natural language and assign the individual steps to their respective equivalents in the flowchart, as well as to the program code.

<b>Gross amount - Algorithm in natural language (stylized prose)</b>

1. Ask user for net amount
2. User enters a net amount
3. If net amount is greater than or equal to 0, go to step 4, otherwise go to step 7
4. Calculate 20% value added tax (VAT) from net amount
5. Calculate the gross amount (sum of net and VAT)
6. Print the VAT and the gross amount to the screen
7. End

<b>Gross amount - Python code</b>

```python
net = float(input("Please enter the net value: "))#Number is read from the console and converted to a float value. It is then stored in the 'net' variable
tax = 0
gross = 0

if net<0: # Check if net is < 0 - If yes, the value is not valid and the tax cannot be calculated
    print("Only positive numbers are allowed!")
else: #If the entered net value is 0 or greater than 0, the tax and gross value are calculated
    tax=net*0.2
    gross=net+tax
    print("Tax is: ",round(tax,2), " Euro")
    print("Gross Value is: ",round(gross,2), " Euro")
```

## Reading Corner - Part 2: if-elif-else Branches


If you want to check multiple conditions, you can use the elif statement in Python. As soon as one of the conditions matches, the entire block is exited. Use the elif statement when only one of the conditions (or none) can be true. 

If you want to include a branch that will be executed if none of the if or elif conditions match, you can use an else statement at the end. The else statement at the end is optional.

The code below will print the current life phase depending on the passed age.

```python
age = 16
if age > 65:            # first condition is checked with if, all others with elif
    print('Pensioner')
elif age > 18:          
    print('Adult')
elif age > 12:          # at this point, the condition is true and the elif block is exited after the print command is executed
    print('Teenager')
elif age > 3:       
    print('Child')
elif age > 0:
    print('Baby')
else:
    print('Invalid age')
```

If several conditions can match, then use multiple if statements. Each condition is checked separately and the corresponding statements belonging to the if block are executed.

The code below will print all life phases the passed age has experienced.

```python
age = 16
if age > 65:        # condition is checked in any case
    print('Pensioner')
if age > 18:        # condition is checked in any case
    print('Adult')
if age > 12:        # condition is checked in any case
    print('Teenager')
if age > 3:         # condition is checked in any case
    print('Child')
if age > 0:         # condition is checked in any case
    print('Baby')
```
