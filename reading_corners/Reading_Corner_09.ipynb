## Reading Corner 1: NumPy II

The following examples show different functions of the NumPy module using an example array.

```python
import numpy as np
```

```python
array = np.random.randint(10, 20, size=(20,))  #Array with random numbers
print('random array:',array)
```

```python
array_slc = array[4:10]  #Index 4 up to 9 is stored in a new array
print(array_slc)

array_slc2 = array[4:15:2]  #Index 4 up to 14 (every second element) is stored in a new array
print(array_slc2)
```

```python
arr_conc = np.concatenate((array_slc, array_slc2))   #Merge 2 arrays
print(arr_conc)

arr_split = np.array_split(arr_conc, 3)     #Split the array into 3 parts
for a in arr_split:
    print(a)
```

```python
arr_ind = np.where(array%2 == 0)     # Select only indexes of the even elements
print(arr_ind[0])
arr_elem = array[array%2 == 0]       # Only even elements
print(arr_elem)
print()

arr2_ind = np.where((array%2 == 0) & (array>15))     # Select only indexes of even elements larger than 15
print(arr2_ind[0])
arr2_elem = array[(array%2 == 0) & (array>15)]       # Only even elements larger than 15
print(arr2_elem)
print()

arr3_ind = np.where((array%2 == 0) | (array>15))     # Select only indexes of even elements or those larger than 15
print(arr3_ind[0])
arr3_elem = array[(array%2 == 0) | (array>15)]       # Only even elements or those larger than 15
print(arr3_elem)
print()
```

## Reading corner 2: Matplotlib

The following program shows how to work with Mathplotlib.


```python
import matplotlib.pyplot as plt
```

```python
data = {'dogs': 22, 'cats': 12, 'hamsters': 80}   #A dictionary as sample data
xpos = [0, 1, 2]    # The position of the individual bars can be set
plt.bar(x=xpos, height=data.values()) #plt.bar for barchart with parameters for the positions of the bars and the data values
plt.xticks(xpos, data.keys()) # Labeling of the axes
plt.suptitle('Pets in the neighbourhood') #A title can also be set for the diagram
plt.show() # The created diagram is printed
```

## Reading Corner 3: Matplotlib basics with numpy
The following program shows how a simple function can be visualized using mathplotlib - the data is stored in a numpy array.

```python
x = np.arange(1,21) #Defining x values (1 to 20)
y = 2*x + 1  #Defining a function y(x)
plt.title("Function (2*x + 1)")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y) #Simple line plot with x and y values can be printed with plt.plot
plt.show()
print('x',x)
print('y',y)
```

## Reading Corner 4: Matplotlib basics with numpy
The following program shows how data can be visualized in a scatterplot. The plot uses two different colors.

```python
weight1=[67,57.2,59.6,59.64,55.8,61.2,60.45,61,56.23,56]    #Numpy arrays serve as data in the plot
height1=[101.7,197.6,98.3,125.1,113.7,157.7,136,148.9,125.3,114.9]
weight2=[61.9,64,62.1,64.2,62.3,65.4,62.4,61.4,62.5,63.6]
height2=[152.8,155.3,135.1,125.2,151.3,135,182.2,195.9,165.1,125.1]
plt.scatter(weight1, height1, color='green') # Define color for first data set
plt.scatter(weight2, height2, color='orange') # If plt.scatter is called twice, two different data sets can be displayed in the same diagram
plt.xlabel('weight', fontsize=10) # Title and font size of x-axis
plt.ylabel('height', fontsize=15)  # Title and font size of y-axis
plt.title('Scatter plot',fontsize=20)   # Title and font size of diagram
plt.show()
```

## Reading corner 5: pandas

The pandas module can be used for data manipulation and analysis, allowing you to work with labeled or relational data in an intuitive way.

The following program shows how data can be processed and analyzed in pandas.

```python
import pandas as pd
```

```python
data = {'Name':['Tom', 'Jack', 'Jane', 'Ricky'],'Age':[28,34,29,42], 'Department':['Tech Support','Sales','Sales','Tech Support'],  'Salary':[4000, 3600,6800,2900]}
# Data sets are stored in a DataFrame.
employee_df = pd.DataFrame(data)

#pandas.read_csv('dateiname.csv') If the data is stored in a CSV file, it can be loaded directly into a DataFrame
```

```python
employee_df.head() # The function head displays the first 5 entries of the DataFrame. The argument n allows any number of entries to be printed
```

```python
employee_df.describe() # describe() calculates different statistical values of the DataFrame
```

```python
df_name_salary=employee_df[['Name','Salary']] # Single or multiple columns of the DataFrame can be selected and stored in a new DataFrame.
df_name_salary.head()
```

```python
employee_df.sort_values(by='Salary', ascending=True).head() # sort_values() allows sorting the DataFrame by a specific column (argument by)
```

```python
employee_df[employee_df.Salary >= 4000].head() # Similar to NumPy, filters can be used to filter specific values in the corresponding columns.
```

```python
grouped_department = employee_df.groupby('Department') # groupby() allows to group the data by one (or more) columns.

grouped_department.sum()
```

```python
grouped_department.describe() # The various pandas functions can now also be applied to the grouped data.
```
