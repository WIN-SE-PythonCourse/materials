## Price calculation - Data Types int and float

So far, we have discussed data types for numbers (*int* for integers, *float* for floating point numbers), as well as the data type string (*str*) for text.

If values are read in by the user, they, by default, have the data type *str*. To be able to use the read-in values in calculations, they must first be converted into numbers (*int* or *float*).

```python
# INPUT
# 3 values are read in by the user and stored in 3 variables
price_per_item = float(input('Price per item: ')) # convert text to float
number_of_items = int(input('Number of items: ')) # convert text to integer
discount = int(input('Discount in percent:'))     # convert text to integer

#PROCESSING
# perform calculations
total_price = price_per_item * number_of_items
print(type(total_price))       #int and float results in float
total_discount = total_price / 100 * discount
new_price = total_price - total_discount

#OUTPUT
print('Total price:', round(total_price, 2))   #Output to console, round rounds the decimal number to the specified digits
print('Discount:', round(total_discount, 2))
print('New price:', round(new_price, 2))
```

## Data Type String

The data type String (*str*) represents text in Python.

To simplify the output of text and values of variables on the console, Python provides an entire set of functions for Strings.

With these functions, Strings can be easily formatted (e.g. case sensitive, or remove spaces), and values of variables can be inserted into strings.

```python
customer_name = 'tommy testperson'
customer_name = customer_name.title() # with the string function 'title', the first letter of a word becomes a capital letter
# Important: calling a string function (like title() in this case) does not change the value of the variable!
# Only with the assignment customer_name = customer_name.title(), the value of the variable customer_name is changed.

print(50* '-')   # By means of '*', strings can be output multiple times
print(f'Customer: {customer_name}') # With 'f', strings can be easily formatted and variable values can be inserted into a string
print('Discount:', round(total_discount, 2), sep = ' --> ')  #sep specifies how the parts of the output are separated from each other
print('New price:', round(new_price, 2), sep = ' --> ')
print('Total price:' + ' ' + str(total_price)) # With +, several strings can be attached to each other (=concatenated)
print(50* '-')
```

# Reading corner: Input & Output - Interaction with the console

There are two important functions in Python for reading in and outputting values from/to the user:

- <font color='green'> input()</font> reads text from the user
- <font color='green'>print()</font> prints text to the user



### input

The <font color='green'> input</font> function always returns a string value. If a number (*int* or *float*) should be read in, it must be converted after being read in.

In the <font color='green'> input</font> function, a prompt for the user can also be specified.

e.g.:

``my_number = int(input('Please enter an integer: '))``

``user_name = input('Please enter a user name: ')``




### print

Prints text, values or variables to the user.
Unlike input, you don't have to convert values here - Python does that for you.

Therefore you can use all types of values and variables (*str*, *int*, *float*, ....) directly in the print function.


e.g.:

``print('Hello')``

``print(25)``

``print(my_number)``

You can also output multiple values or variables one after the other in a <font color='green'> print</font> statement (in the same line).

Alternatively, you can use f Strings.

``print('3+4 =',7)``

``print(f'The number you entered is: {my_number}')``

```python
# INPUT

# Read values from the user:

## this input asks the user for the age and converts the entered value directly to int
my_age = int(input('Please enter your age: '))


## this input asks the user for the body height and converts the entered value directly to float
my_size = float(input('Please enter your size (in m): '))


## this input asks the user for the name - Since every input is automatically a string, nothing needs to be
## converted here
my_name = input('Please enter your name: ')


# After the values entered by the user have been stored in the respective variables,
# these values can be further used (e.g. for calculations)


#OUTPUT

# The print function can be used to output information to the user.
print('') # A print without text outputs an empty line
print('This is me:')

print('Hello, my name is', my_name)
print('I am ', my_age, 'years old')
print(f'I am {my_size}m tall ')

print('')
print('-'*35)

# With the type function you can determine which data type a corresponding variable has:
print('')
print('Types of my variables:')

print('The type of the variable "my_name" is:', type(my_name))
print('The type of the variable "my_age" is:', type(my_age))
print('The type of the variable "my_size" is:', type(my_size))
```
