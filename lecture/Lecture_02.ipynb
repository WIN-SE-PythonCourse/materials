## Strings

### String Functions

```python
name = 'reinhold ploesch' ## String are declared using '' or ""
another_name = "Iris Groher"

print(name)

print('reinhold\nploesch') # \n represents a 'special character' for new line
```

```python
text = '      Hello World     '
```

```python
print(text.strip()) # The function strip removes whitespaces from a string
print(text.rstrip())
print(text.lstrip())
```

```python
print('hallo'.upper())  # The function upper converts all characters to upper case

print('Reinhold Ploesch'.lower()) # The function upper converts all characters to upper case
print(name.title()) # The function title converts the first character of each word of the string to upper case
print(name)
name = name.title()
```

#### Check Strings

```python
print(name.isnumeric()) ## checks if the String only contains numbers
print('444'.isnumeric())
```

#### Len and Find

```python
name = 'Franz'
```

```python
print(len(name)) # len() returns the length of a String = the number of characters (including whitespaces)
print(len('Hello World      '))

print(name.find('a')) # Find searches for a character, or string and returns the index of its first occurence
print(name.find('anz'))
print(name.find('x')) # If the string does not contain the character you are searching for, Python returns -1
```

### String concatenation

```python
first_name = 'iris'
last_name = 'groher'
```

```python
print('-'*25) # The * operator used with a string prints the String x times.
print(first_name + ' ' + last_name) # Strings can be concatenated (joined togehter) using the + operator
print(first_name, last_name)
print('-------------------------')
```

### String Formatting

```python
name = f' Hello {first_name} {last_name}' # The f-String allows formatting strings more easily without the + operator
print(name)
```
