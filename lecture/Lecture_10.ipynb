### CSV Module

```python
import csv   #import csv module
```

#### Reading CSV Files

```python
with open('data.csv') as csvfile:    #open csv file (download file from moodle)
    csvreader = csv.reader(csvfile,delimiter=',')  #csv reader object
    for row in csvreader:
        print(row[1])
        print(row)                  #print the csv file row by row (each row is a list of strings, including the header)
```

```python
spreadsheet = []
with open('data.csv') as csvfile:
    csvreader = csv.DictReader(csvfile,delimiter=',')   #reads csv file as a list of dictionaries
    for row in csvreader:           #each row is a dictionary (key is column name, value is value in row)
        spreadsheet.append(row)

print(spreadsheet)         #print list of dictionaries
print()

for entry in spreadsheet:       #traverse all dictionaries in the list
    print(entry['BundeslandID'])   #print all entries with key 'Bundesland'
```

```python
data = []
with open('data.csv') as csvfile:
    reader = csv.reader(csvfile,delimiter=',')
    for row in reader:        #each row is a list of Strings
        data.append(row)      #data is a list of lists (2-dimensional array)

print(data)                   #print list of lists
print()

for i in range(len(data)):          #traverse all lists in data
    for j in range(len(data[i])):   #for each list traverse all elements
         print(data[i][j])          #print each element
```

#### Writing CSV Files

```python
with open('data.csv', 'a') as csvfile:
    writer = csv.writer(csvfile,delimiter=',')    #csv writer object
    writer.writerow(['26.02.20', 'Salzburg', '5', '560643'])   #add rows to csv file
    writer.writerow(['26.02.20', 'Tirol', '6', '760161'])

with open('data.csv') as csvfile:    #open csv file
    csvreader = csv.reader(csvfile,delimiter=',')  #csv reader object
    for row in csvreader:
        print(row)
```

```python
data= [{'Datum':'26.02.20', 'Bundesland':'Salzburg', 'BundeslandID':'5', 'AnzEinwohner':'560643'}, {'Datum':'26.02.20', 'Bundesland':'Tirol', 'BundeslandID':'6', 'AnzEinwohner':'760161'}]
with open('data-new.csv', 'w') as csvfile:    #a new csv file is created
    fieldnames = ['Datum', 'Bundesland', 'BundeslandID', 'AnzEinwohner']   #header is a list of strings
    writer = csv.DictWriter(csvfile, fieldnames = fieldnames)    #writer for writing dictionaries to the csv file
    writer.writeheader()                                         #write header to the file
    writer.writerow(data[0])    #write rows
    print(data[0])
    writer.writerow(data[1])
    print(data[1])
```

### NumPy Module

```python
import numpy as np      #import numpy module and use alternative name (np)
```

#### Creating NumPy arrays

```python
ar = np.array([1,2,3,4])    #create numpy array with array function
ar2 = np.array([1.5,-2.2,3.7,-4.1])

print(ar)
print(ar2)
print(type(ar))
```

```python
ar_dim0 = np.array(1)    #array with dimension 0
ar_dim1 = np.array([1,2,3,4])   #array with dimension 1
ar_dim2 = np.array([[1, 2, 3], [4, 5, 6]])  #array with dimension 2 (matrix)

print(ar_dim0)
print()
print(ar_dim1)
print()
print(ar_dim2)
```

```python
array = np.zeros(5)     #create array with 5 elements (all 0)
array2 = np.ones(2)     #create array with 2 elements (all 1)

print(array)
print(array2)

array = np.arange(4, 20)   #create array with values from a to 19
print(array)

array = np.random.rand(5)   #create array with 5 elements (random numbers between 0 and 1)
print(array)
```

#### Accessing NumPy arrays

```python
ar_dim1 = np.array([1,2,3,4])
ar_dim2 = np.array([[1, 2, 3], [4, 5, 6]])

print(ar_dim1[0], '-->', ar_dim1[-1])    #first and last element
print(ar_dim2[0, 0], '-->', ar_dim2[0, -1])   #first array first element, first array last element
print(ar_dim2[1, 0], '-->', ar_dim2[1, -1])   #second array first element, second array last element
```

#### CSV to NumPy

```python
data = []
with open('data.csv') as csvfile:    #open and read csv file
    csvreader = csv.reader(csvfile,delimiter=',')
    for row in csvreader:
        data.append(row)

np_data = np.array(data)     #store data in numpy array
print(np_data)               #print numpy array
```
