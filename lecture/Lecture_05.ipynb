## Lists

```python
shopping_list = ['Programming with Python', 'Cat Food', 'Dog Food', 'Shoes']  #list with 4 string elements
prices = [50, 24, 19.90, 27]        #list with 4 numeric elements
mixed_list = ['String', 25, 'another String' , 'abcd', 26, ['listinalist', 5]]   #list with elements of different types, also contains a list
empty_list = []   #empty list

print(shopping_list)   #print(list) prints the contents of the list
print(prices)
print(mixed_list)
print(empty_list)
```

```python
first_elem = shopping_list[0]  #access first element of the list (index starts with 0)
print(first_elem)

last_elem = shopping_list[-1]   #access last element of the list (last index is -1)
print(last_elem)
```

```python
shopping_list[0] = 'Programming with C++'  #changing the first element in the list
print(shopping_list)

shopping_list[-1] = 'Salad'    #changing the element at index 4
print(shopping_list)
```

```python
shopping_list[5] = 'IPhone 12 Pro'    #accessing an index that does not exist results in an error
```

## Dictionary

```python
shopping_cart = { 'Programming with Python' : 49.90 , 'Cat Food' : 5 , 'Dog Food' : 3  } # A dictionary can be created with key,value pairs

print(shopping_cart)
```

```python
my_dict = {} ## {} creates an empty dictionary that can be filled with values later on

print(my_dict)
```

```python
my_dict['zahl1'] = 25 # Dictionary are always accessed via the respective key - they do not have an index
my_dict['zahl2'] = 7

print(my_dict)

print(my_dict['zahl1'])

print(my_dict['zahl7']) # When accessing an element, the key must exist in the dictionary! - Otherwise an error is thrown


```

```python
shopping_cart = { 'Programming with Python' : 49.90 , 'Cat Food' : 5 , 'Dog Food' : 3  }

shopping_cart['Milk'] = 1.5 ## Values can be added to the dictionary by adding the key with the respective value

print(shopping_cart)

shopping_cart['Milk'] = 0.75 # A value can only exist once, ther are no duplicates in a dictionary
print(shopping_cart) # If a value is set the second time, the original value is replaced

shopping_cart.update({'Milk': 1.75})
print(shopping_cart)

shopping_cart.pop('Milk')
print(shopping_cart)

if 'Cat Food' in shopping_cart: # In checks if a certain key is part of the dictionary
    shopping_cart.pop('Cat Food')

print(shopping_cart)


```

```python
for item in shopping_cart: #  With item in dict, the for loop iterates over the keys
    print(item)
```

```python
for item in shopping_cart.values(): #  With item in dict.values(), the for loop iterates over the values
    print(item)
```

```python
for k,v in shopping_cart.items(): #  With key,value in dict.items(), the for loop can access both the key and the value of an item    sum += v
    print(k,v)
```

## Exam Results

Create a Python program that reads exam results (points) from the console and stores them in a list.

Users can enter as many points until they enter -1 which ends the program and prints the list of exam results.


```python
exam_results = [] # Create a new empty list

value = int( input(f'Points achieved by student 1: ') )
while value != -1 :
    exam_results.append(value) # Add each value to the list
    value = int( input(f'Points achieved by student {len(exam_results)+1}: ') )


for res in exam_results: # Iterate over the list and print each value
    print(res)


```

## Person Dictionary


Create a Python program that represents a person in a dictionary with
first name: John
last name: Doe
age: 52
spouse: Sarah
children: Linda, Fred

The program prints the information about the person.


```python
person = {} # Create a new empty dictionary

person['fname'] = 'John' # Add the elments as key,value pairs to the dictionary
person['lname'] = 'Doe'
person['age'] = 52
person['spouse'] ='Sarah'
person['children'] = ['Linda','Fred']


for k,v in person.items(): # items() iterates over the dictionary and provides both, the key (stored in the variable k) and the value (stored in the variable v).
    print(k,v)
```
