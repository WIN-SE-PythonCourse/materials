## Importing modules

```python
import random # Import the module called 'random'

random_number = random.randint(12,15) #call the function randint from this module

print(random_number)
```

```python
import os # Import the module called 'os'

print(os.listdir())
```

```python
from math import pow # Import a specific function (pow) from the math module

result =pow(7,2) # pow can now be called without the modulename (math.)
print(result)
```

```python
from math import * # Import all functions from the math module module

result = pow(7,2) # pow can now be called without the modulename (math.)
result2 = sqrt(64) # sqrt can now be called without the modulename (math.) ...
result3 = floor(26.8554)
result4 = ceil(23.334)

print(result4)
```

```python
from math import log10 as logarithm  # Import a specific function from the math module, but give it a different name

result = logarithm(100) # The log10 function (originally called log10 in the math molule) was imported with the new name logarithm
print(result)
```

## Example: Random Person


For the Python exercise students should present their solutions in class.
Write a Python program that reads all student names from a text file, stores them in a list, and then randomly selects 5 students that should present one exercise tasks.


```python
import random
```

```python
with open('student_list.txt') as name_file:
    students =name_file.read().splitlines()
```

```python
for s in students:
    print(s)
```

```python
## Variant 1: random.randint:

for i in range(5):
    selected = random.randint(0,len(students)-1)
    print(f'Selected student {selected}: {students[selected]}')
```

```python
## Variant 2: random.choices:

selected_student= random.choices(students)
print(f'Selected student: {selected_student}')
```

```python

#Select more than 1 student at the same time - choices has a default argument k
#(number of choices - which is 1 by default)
selected_student= random.choices(students, k=5)
print(f'Selected student: {selected_student}')
```
