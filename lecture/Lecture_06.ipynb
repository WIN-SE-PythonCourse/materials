## Multiple Return Values

```python
def min_max_price(list):
    min_price = min(list)
    max_price = max(list)
    return min_price, max_price
```

```python
shopping_cart = {'Programming with Python Book':49.90 , 'Cat Food':3, 'Apple IPhone 11Pro':1299}

## grab all values fromn the dictionary and convert them into a list
prices = list(shopping_cart.values())
print(prices)
```

```python
### First Option with a tuple


min_max_result = min_max_price(prices)

print(type(min_max_result))

print(f'Minimum:{min_max_result[0]}')
print(f'Maximum:{min_max_result[1]}')
```

```python
### Second Option with a multiple variables

min_result, max_result = min_max_price(prices)

print(type(min_result))

print(f'Minimum:{min_result}')
print(f'Maximum:{max_result}')
```

## Rectangle

Write a python program with a function read_side_lengths that reads the side lengths of a rectangle.
Write another function area_and_perimenter that calculates the area and the perimeter of the rectangle. The function takes as optional arguments the side lengths of the rectangle.

```python
def read_side_lengths():            #Definition of function with no arguments
    a = float(input('side a: '))    #Read side a from the user
    b = float(input('side b: '))    #Read side b from the user
    return a, b                     #Return 2 values (side a and side b) to the caller
```

```python
def area_and_perimeter(a=1, b=1):   #Definition of function with two arguments, both with default values
    area = a * b                    #Calculate area
    perimeter = (a + b) * 2         #Calculate perimeter
    return area, perimeter          #Return 2 values (area and perimeter) to the caller
```

#### With individual variables:

```python
a, b = read_side_lengths()                     #Save the 2 return values in 2 variables
area, perimeter = area_and_perimeter(a, b)     #Save the 2 return values in 2 variables
ap = area_and_perimeter()                      #Save the 2 return values in 1 tuple


print('-'*15)
print('area:', area)
print('perimeter:', perimeter)
print('area:', ap[0])                          #Access first element of the tuple (area)
print('perimeter:', ap[1])                     #Access second element of the tuple (perimeter)
```

#### With a tuple:

```python
a_b = read_side_lengths()                      #Save the 2 return values in 1 tuple
area, perimeter = area_and_perimeter(a_b[0], a_b[1])      #Access elements of the tuple
print('area:', area)
print('perimeter:', perimeter)
print('-'*15)
```

#### With default Values

```python
a_b = read_side_lengths()
area, perimeter = area_and_perimeter(a_b[0])     #Call function with 1 argument (the second one uses the default value)
print('area:', area)
print('perimeter:', perimeter)
```

```python
def read_results():
    print('reading values...')
    exam_results = [] # Create a new empty list
    value = int( input(f'Points achieved by student 1: ') )
    while value != -1 :
        exam_results.append(value) # Add each value to the list
        value = int( input(f'Points achieved by student {len(exam_results)+1}: ') )

    return exam_results
```

```python
def print_results(results_to_print):
    print('printing results...')
    for res in results_to_print: # Iterate over the list and print each value
        print(res)
```

```python
my_exam_data = read_results()

print('-'*15)

print_results(my_exam_data)
```
